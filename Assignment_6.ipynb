1. What are escape characters, and how do you use them?

Answer:

The backslash '\' is a special character also called the escape character.prefixing this special character with '\' turns it into an ordinary character.

examples:

a.Single quote 

test = 'it\n's fine'

print(txt)
output : it's fine

b . back slash

test = 'it\\n's fine'
print(txt)
output : it'\s fine

c.New line
test = 'Hello\n world'
print(txt)
output : 
Hello 
world
